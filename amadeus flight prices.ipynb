```python
results = []

for offer in loopData['data']:
    entry = {
        'id': offer.get('id'),
        'seats': offer.get('numberOfBookableSeats'),
        'duration': offer['itineraries'][0]['duration'],
        'price_total': offer['price']['total'],
        'price_base': offer['price']['base'],
        'currency': offer['price']['currency'],
        'traveler_type': offer['travelerPricings'][0]['travelerType'],
        'fare_class': offer['travelerPricings'][0]['fareDetailsBySegment'][0]['class'],
        'cabin': offer['travelerPricings'][0]['fareDetailsBySegment'][0]['cabin'],
        'segments': len(offer['itineraries'][0]['segments']),
    }
    results.append(entry)
```
